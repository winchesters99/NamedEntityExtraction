```python
import pandas as pd
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
#print(soup.prettify())
```

```python
My_table = soup.findAll('div',{'class':'div-col columns column-width'})
My_table
```

```python
s=""
for i in My_table:
    s+=str(i)
s
```

```python
a=s.split('<li>')
q1=[]
q2=[]
df1 = pd.DataFrame()

for i in a:
    if "href" not in i:
        continue
    elif "mw-redirect" in i:
        q=i.split('"')
        q1.append(q[3])
        q2.append(q[5])
        
    else:
        q=i.split('"')
        q1.append(q[1])
        q2.append(q[3])
        #print(q[1],"\t",q[3])
df1['Actor'] = q2
df1['PageLink'] = q1
df1
df1.to_csv('output.csv')
```
