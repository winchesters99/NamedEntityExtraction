```python
import requests
import csv
from bs4 import BeautifulSoup
import pandas as pd

WIKI_URL = "https://en.wikipedia.org/wiki/List_of_towns_in_India_by_population"
""
req = requests.get(WIKI_URL)
soup = BeautifulSoup(req.content, 'lxml')
table_classes = {"class":"sortable"}
wikitables = soup.findAll("table", table_classes)

output_rows = [['City','State','PageLink']]

for table in wikitables:
	j = 0
	for table_row in table.findAll('tr'):
		
		columns = table_row.findAll('td')
		
		output_row = []
		
		i = 0
		
		for column in columns:
			
			if i == 0:
				#j += 1            
				output_row.append(j)
				output_row.append(column.string)
				
			#if j == 0:
				#output_row.append(column.string)
			elif i == 1:
				titleAndLink = column.find('a')
				output_row.append(titleAndLink['title'])
				output_row.append(titleAndLink['href'])
				
			else:
				break

			i += 1
			#j += 1
		output_rows.append(output_row)
		
		#column = columns[0].string
		#output_row.append(column)

		#column = columns[1].b.title
		#output_row.append(column)

		#column = columns[1].b.a
		#output_row.append(column)

		#output_rows.append(output_row)

for row in output_rows:
	print(row)

#with open('output.csv', 'wb') as csvfile:
    #writer = csv.writer(csvfile)
    #writer.writerows(output_rows)

with open('output_town.csv', mode='w') as output_file:
    employee_writer = csv.writer(output_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for row in output_rows:
        employee_writer.writerow([row])


```
