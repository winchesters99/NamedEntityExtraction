```python
import re
import requests
import csv
import time
import threading
import nltk
nltk.download('punkt')
from bs4 import BeautifulSoup as soup
from nltk.tokenize import sent_tokenize, word_tokenize

```

```python
#GLOBAL VARIABLES
filename = "output1new.csv"
curr_row = []
link_prefix = 'http://www.wikipedia.org'
colTitles = []
```

```python
def FetchEntityContent(articleSoup,entityName):

    list_of_all_p_tags = articleSoup.find_all('p')
    
    #print(list_of_all_p_tags)

    list_of_sent_cont_entity = []
    list_of_sent_cont_entity_2 = []
    
    #z=0
    for i in list_of_all_p_tags:
        
        #z+=1
        #print(i)
        
        #getting the paragraph text
        
        paragraph_text = i.get_text()
        
        #print(paragraph_text)
    
        list_of_sentences = sent_tokenize(paragraph_text)
    
        #print("RAW DATA:")
        #print(list_of_sentences)
    
        #for j in list_of_sentences:
        #print('########################################################################################')
        #print(j)
    
    
        #first clean the entity of spaces,then split
    
        entity_parts_list = entityName.split(" ")
        
        #print(z)
        #print(entity_parts_list)
    
        #either search for the entire entity or parts(but not initial letters)
    
        entity_parts_list_final = (entityName,entity_parts_list[-1])
        #print(entity_parts_list_final)
    
    
        #search for entity in the text
        for k in list_of_sentences:
            for j in entity_parts_list_final:
                if re.search(j,k):
                    list_of_sent_cont_entity.append(k)
                    break
                else:
                    continue
        
        #list_of_sent_cont_entity_2 = Remove(list_of_sent_cont_entity)
    f1 = entityName+'.txt'
    f1=open(f1, "a+")
    for j in list_of_sent_cont_entity:
        #print('########################################################################################')
        a='\['
        b='\]'
        a=a+".*?"+b
        j=re.sub(a,"",str(j))
        f1.write(str(j))
        print(j)
```

```python
def FileParser():
    
    with open(filename, 'r') as csv_file:
        
        csvreader = csv.reader(csv_file,delimiter = ',')

        i = 0
        for row in csvreader:
            
            if i==0:
                i+=1
                continue
            curr_row = row
            entityName = str(curr_row[1])
            link = link_prefix + str(curr_row[2])
            
            #PRINT STATEMENTS FOR DEBUGGING

            print(i)
            print('Entity : ',entityName)
            print('Wikipedia Link : ',link)
            
            #Entity and link extracted,now scrape contextual data

            article = requests.get(link)
            
            articleSoup = soup(article.content,'html.parser')
            print(articleSoup.title)
            
            
            
            
            
			#link isthe request library http obj
            
            #FetchEntityContent(articleSoup,entityName)

            t2 = threading.Thread(target=FetchEntityContent, args=(articleSoup,entityName,))

            t2.start()

            t2.join()
            
            i += 1
```

```python
if __name__ == "__main__":
    
    print("Starting Execution:")
    
    t1 = threading.Thread(target=FileParser) 
    
    t1.start()

    t1.join()
```
